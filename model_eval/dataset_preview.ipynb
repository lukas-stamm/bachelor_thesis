```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Path to the base dataset directory
base_path = '../data'

# List to collect (image_path, label, source_dataset)
all_images = []

# Traverse through datasets
for dataset_name in ['DALLE_dataset', 'IMAGEN_dataset', 'SD_dataset']:
    for label in ['real', 'fake']:
        dir_path = os.path.join(base_path, dataset_name, label)
        if os.path.exists(dir_path):
            for file in os.listdir(dir_path):
                if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                    all_images.append((
                        os.path.join(dir_path, file),
                        label,
                        dataset_name.replace('_dataset', '')
                    ))

# Randomly select 5 images
sample_images = random.sample(all_images, 5)

# Plotting
plt.figure(figsize=(15, 5))
for i, (img_path, label, dataset) in enumerate(sample_images):
    img = Image.open(img_path)
    plt.subplot(1, 5, i + 1)
    plt.imshow(img)
    plt.title(f"{label}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
